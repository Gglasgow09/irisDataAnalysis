```python

import matplotlib.pyplot as plt

# Assuming df is your DataFrame
sepal_length = df['SepalLengthCm']
petal_length = df['PetalLengthCm']

fig, ax = plt.subplots(figsize=(8, 6))

ax.scatter(sepal_length, petal_length, color='blue', alpha=0.5)

ax.set_xlabel('Sepal Length (cm)')
ax.set_ylabel('Petal Length (cm)')
ax.set_title('Relationship between Sepal Length and Petal Length')

plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='SepalLengthCm', y='PetalLengthCm', hue='Species')

plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.title('Relationship between Sepal Length and Petal Length')

plt.legend(title='Species')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

y = df[['SepalLengthCm']]
x = df[['SepalWidthCm']]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

print("x_train:")
print(x_train.head())
print("\nx_test:")
print(x_test.head())
print("\ny_train:")
print(y_train.head())
print("\ny_test:")
print(y_test.head())
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lr = LinearRegression()
lr.fit(x_train, y_train)

y_pred = lr.predict(x_test)

print("First five records from actual (y_test) and predicted (y_pred) objects:")
actual_vs_predicted = pd.DataFrame({'Actual': y_test.values.flatten(), 'Predicted': y_pred.flatten()})
print(actual_vs_predicted.head())

mse = mean_squared_error(y_test, y_pred)
print("\nMean Squared Error:", mse)
```
